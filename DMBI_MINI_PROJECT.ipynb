**DATA CLEANING AND REMOVE NaN VALUES:**

**Fill Missing Value in Year_of_Release Feature**

```python
df['Year_of_Release'] = df['Year_of_Release'].fillna(df['Year_of_Release'].mean())
# Change Year dtype to int32
df['Year_of_Release'] = df['Year_of_Release'].astype('int32')
```

**Fill missing value in Publisher Feature**

```python
df.Publisher.value_counts(normalize=True)
```

```python
df['Publisher'] = df['Publisher'].fillna(df['Publisher'].mode()[0])
```

**Fill missing value in Critic_Score and User_Score Feature**

```python
df['Critic_Score'] = df['Critic_Score'].fillna(df['Critic_Score'].mean())
df['Critic_Score'] = df['Critic_Score'].astype('int32')
```

```python
df['User_Score'] = df['User_Score'].fillna(df['User_Score'].mean())
df['User_Score'] = df['User_Score'].round(decimals=1)
```

**Fill missing value in Critic_Count and User_Count Feature**

```python
df['Critic_Count'] = df['Critic_Count'].fillna(df['Critic_Count'].median())
df['Critic_Count'] = df['Critic_Count'].astype('int32')
```

```python
df['User_Count'] = df['User_Count'].fillna(df['User_Count'].median())
df['User_Count'] = df['User_Count'].astype('int32')
```

**Fill missing value in Developer and Rating Feature**

```python
df['Developer'] = df['Developer'].fillna(df['Developer'].mode()[0])
```

```python
df['Rating'] = df['Rating'].fillna(df['Rating'].mode()[0])
```

**DATA MODELING:**

Implementing LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
data = df.copy()
le = LabelEncoder()
feature = ["Platform", "Genre", "Publisher", "Developer", "Rating"]
for col in feature:
    data[col] = le.fit_transform(df[col])
```

```python
from sklearn.model_selection import train_test_split

X = data[['Platform', 'Genre','Publisher','Developer','Rating','NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].values
y = data['Global_Sales'].values

from sklearn import preprocessing
y = preprocessing.LabelEncoder().fit_transform(y)
y = np.array(y).reshape(-1,1)
print(X.shape)
print(y.shape)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=45)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

**1. LINEAR REGRESSION:**

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split


# Import r2 score for Calculation
from sklearn.metrics import r2_score
lr = LinearRegression()

lr.fit(X_train, y_train)

pred = lr.predict(X_test)

r2_MultiLinear = r2_score(y_test,pred)
print(r2_MultiLinear)

print(lr.score(X_test, y_test))
```

**2. XGB REGRESSOR:**

```python
from xgboost import XGBRegressor
xgb = XGBRegressor(n_estimators=500,max_depth=5,booster='gbtree',n_jobs=-1,learning_rate=0.1,reg_lambda=0.01,reg_alpha=0.3)
xgb.fit(X_train, y_train)

pred = xgb.predict(X_test)

print(r2_score(y_test, pred))
```

**3. RIDGE CV:**

```python
from sklearn.linear_model import RidgeCV
ridge_model = RidgeCV(scoring='r2', alphas=[0.001,0.005,0.001,0.005,0.01,0.1,1.0,10], cv=5)
ridge_model.fit(X_train,y_train)
y_train_pred = ridge_model.predict(X_train)
y_test_pred = ridge_model.predict(X_test)
print(r2_score(y_true=y_train,y_pred=y_train_pred))
print(r2_score(y_true=y_test,y_pred=y_test_pred))
```

**4. LASSO CV:**

```python
from sklearn.linear_model import LassoCV
lasso_model = LassoCV(alphas=[0.001,0.005,0.001,0.005,0.01,0.1,1.0,10], cv=5)
lasso_model.fit(X_train,y_train)
y_train_pred = lasso_model.predict(X_train)
y_test_pred = lasso_model.predict(X_test)
print(r2_score(y_true=y_train,y_pred=y_train_pred))
print(r2_score(y_true=y_test,y_pred=y_test_pred))
```

**5. ADABOOSTCLASSIFIER:**

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

# y_train = LabelEncoder().fit_transform(y_train)
model = AdaBoostClassifier(n_estimators=100,learning_rate=1,random_state=123)
model.fitted = model.fit(X_train, y_train.ravel())
model.predictions = model.fitted.predict(X_test)

print(confusion_matrix(y_test, model.predictions))
print(accuracy_score(y_test, model.predictions))
```
